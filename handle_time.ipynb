```python
dataHours = data.loc[:,["ID", "Opening Hours"]]
dataHours
```

```python
dataHours = dataHours.fillna(method='ffill')
dataHours
```

```python
def convertToMinutes(data):
    newData = data.split(':')
    return int(newData[0]) * 60 + (newData[1]) + 1
```

```python
header = ['ID', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
listTimes = []
for index in range(len(dataHours)):
    openingHours = dataHours.iloc[index][1]
    ID = dataHours.iloc[index][0]
    newItem = [ID]
    if isinstance(openingHours, str):
        items = openingHours.strip("']['").split("', '")
        for item in items:
            item = item.encode('ascii', 'ignore').decode('unicode_escape')
            listInfo = item.split(": ")
            day = listInfo[0]
            times = listInfo[1].strip()
            timeOpen = 0
            timeClose = 0
            if times == "Open 24 hours":
                timeOpen = '00:00 AM'
                timeClose = '11:59 PM'
                times = timeOpen + '  ' + timeClose
            else:
                times = re.sub('[\u202f\u2009]+', ' ', times)
            
            newItem.append(times)
    listTimes.append(newItem)
        
```

```python
with open('/home/gracie/Documents/DaNaTravel/training/data/opening_hours.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerows(listTimes)
```
