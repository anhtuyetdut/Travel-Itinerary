```python
import csv
import pandas as pd
import numpy as np
```

```python
def readCsv(name):
    data = pd.read_csv('/home/gracie/Documents/DaNaTravel/training/data/{}.csv'.format(name))
    print(len(data))
    data.head(10)
    return data
```

```python
def removeEmptyRating(data):
    newData = data.dropna(how='any',subset=['Rating'])
    print(len(newData))
    newData.head(10)
    return newData
```

```python
def removeUnvalidRating(data):
    dataChange = data
    dataChange['Rating'] = np.where(data['Rating'] < 3, 0, data['Rating'])
    newData = dataChange[dataChange.Rating != float(0)]
    print(len(newData))
    newData.head(10)
    return newData
```

```python
def handleUserRatingTotal(data):
    newData = data[data['User Ratings Total'] >= 20]
    return newData
```

```python
def resetIndex(data):
    newData = data
    newData.reset_index(inplace=True, drop=True)
    return newData
```

```python
def handle_types(data):
    valid_types = ['lodging', 'point_of_interest', 'establishment', 'tourist_attraction', 'food', 'cafe', 'restaurant']
    invalid_types = ['point_of_interest', 'establishment']
    listRowRemove = []

    for i in range(len(data)):
        row = data.iloc[i]
        types = row['Types'].strip("']['").split("', '")
        if all(item in invalid_types for item in types) and len(types) == 2:
            listRowRemove.append(i)
        elif all(item in valid_types for item in types) == False:
            listRowRemove.append(i)

    newData = data.drop(listRowRemove)
    return newData
```

```python
def write_csv(data):
    data.to_csv('/home/gracie/Documents/DaNaTravel/training/data/newData.csv', mode='a', index=False)
```

```python
def write_header():
    header = ['ID', 'Name', 'Address', 'Latitude', 'Longitude', 'Opening Hours', 'Types', 'Rating', 'User Ratings Total']
    with open('/home/gracie/Documents/DaNaTravel/training/data/newData.csv', 'a', encoding='UTF8', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(header)
```

```python
fileNames = ['dong', 'nam', 'bac']
write_header()
for name in fileNames:
    print(name)
    data = readCsv(name)
    data = removeEmptyRating(data)
    data = removeUnvalidRating(data)
    data = handleUserRatingTotal(data)
    data = resetIndex(data)
    newData = handle_types(data)
    write_csv(newData)
```

```python
def handleDuplicates(data):
    data = data.sort_values(by=['ID'])
    data = data.drop_duplicates(subset=['ID'], keep='last')
    return data
```
