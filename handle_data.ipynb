```python
import csv
import pandas as pd
import numpy as np
import math
```

```python
def readCsv(name):
    data = pd.read_csv('/home/gracie/Documents/DaNaTravel/training/data/data1/{}.csv'.format(name))
    print(len(data))
    data.head(10)
    return data
```

```python
def removeEmptyRating(data):
    # newData = data.dropna(how='any',subset=['Rating'])
    # print(len(newData))
    # newData.head(10)
    df_empty = pd.DataFrame(columns=['ID', 'Name', 'Address', 'Latitude', 'Longitude', 'Opening Hours', 'Types', 'Rating', 'User Ratings Total'])
    df = pd.DataFrame(columns=['ID', 'Name', 'Address', 'Latitude', 'Longitude', 'Opening Hours', 'Types', 'Rating', 'User Ratings Total'])
    for i, row in data.iterrows():
        new_df = pd.DataFrame([row])
        if math.isnan(float(row['Rating'])): df_empty = pd.concat([df_empty, new_df])
        else: df = pd.concat([df, new_df])
    return df, df_empty
    # return newData
```

```python
def removeUnvalidRating(data):
    dataChange = data
    dataChange['Rating'] = np.where(data['Rating'] < 3, 0, data['Rating'])
    newData = dataChange[dataChange.Rating != float(0)]
    print(len(newData))
    newData.head(10)
    return newData
```

```python
def handleUserRatingTotal(data):
    newData = data[data['User Ratings Total'] >= 20]
    return newData
```

```python
def resetIndex(data):
    newData = data
    newData.reset_index(inplace=True, drop=True)
    return newData
```

```python
def handle_types(data):
    valid_types = ['food', 'cafe', 'restaurant', 'tourist_attraction', 'amusement_park', 'park', 'church', 'natural_feature', 'museum']
    invalid_types = ['point_of_interest', 'establishment']
    listRowRemove = []

    for i, row in data.iterrows():
        types = row['Types'].strip("']['").split("', '")
        if any(item in valid_types for item in types) == False:
            listRowRemove.append(i)

    newData = data.drop(listRowRemove)
    return newData
```

```python
def handleDuplicates(data):
    data = data.sort_values(by=['ID'])
    data = data.drop_duplicates(subset=['ID'], keep='last')
    return data
```
